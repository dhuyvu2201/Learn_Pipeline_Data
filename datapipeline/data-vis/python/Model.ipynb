```python
def processing_data(data):
    label_encoder = LabelEncoder()
    df=data.copy()
    object_columns = ['location','description','current_weather_ic','current_pollution_maincn','current_pollution_mainus']
    for col in object_columns:
        df[col] = label_encoder.fit_transform(df[col])    
    df = df.drop(columns=['day'])
    return df
df = processing_data(df_merge)
df.head()
```

```python
def plot_features(df, feature_names, plot_type='distplot'):
    num_features = len(feature_names)
    grid_size = (2, 3)
    num_rows = (num_features - 1) // grid_size[1] + 1
    num_cols = min(num_features, grid_size[1])
    plt.figure(figsize=(15, 5 * num_rows))

    for i, col in enumerate(feature_names, 1):
        plt.subplot(num_rows, num_cols, i)
        if plot_type == 'boxplot':
            sns.boxplot(x=col, data=df)
        elif plot_type == 'scatter':
            sns.scatterplot(x=col, y='Target_Column', data=df)  # Chọn 'Target_Column' là cột dữ liệu cần so sánh
        else:
            sns.violinplot(x=col, data=df)

        plt.title(f"{plot_type.capitalize()} of {col}", fontsize=14)
        plt.xlabel(col, fontsize=12)

    plt.tight_layout()
    plt.show()
```

```python
def Iso_handle(data, target):
    df = data.copy()
    df = df.drop(columns='forecastdate')
    df.dropna(inplace=True)

    model = IsolationForest(n_estimators=300, max_samples='auto', contamination=0.1, max_features=1.0)
    model.fit(df.drop(columns=target))
    outliers = model.predict(df.drop(columns=target))
    print('Total outliers =', sum(outliers < 0))  

    new_df = df[outliers > 0]
    new_df = data[data.index.isin(new_df.index)]
    new_df = new_df.copy()
    
    new_df['forecastdate'] = pd.to_datetime(new_df['forecastdate'])
    #ew_df['day'] = new_df['forecastdate'].dt.date
    new_df['hour'] = new_df['forecastdate'].dt.hour
    new_df = new_df[[col for col in new_df.columns if col != 'forecastdate'] + ['forecastdate']]
    return new_df
```

```python
def split_data(data, target):
    df = data.copy()
    X = df.drop(columns=target).values
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2023)
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train[:, :-1])
    X_test_scaled = scaler.transform(X_test[:, :-1])

    timestamp_train = X_train[:, -1].reshape(-1, 1)
    timestamp_test = X_test[:, -1].reshape(-1, 1)
    
    X_train_with_time = np.concatenate((X_train_scaled, timestamp_train), axis=1)
    X_test_with_time = np.concatenate((X_test_scaled, timestamp_test), axis=1)
    
    return X_train_with_time, X_test_with_time, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_data(df_1,'description')
```

```python
from sklearn.metrics import accuracy_score

def evaluate_model(y_true, y_pred):
    report = classification_report(y_true, y_pred)
    print("Classification Report\n", report)
    cm = confusion_matrix(y_true, y_pred)
    #sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    #plt.show()
    accuracy = accuracy_score(y_true, y_pred)
    return accuracy

def Build_Model(X_train, X_test, y_train, y_test, model):
    pipeline = make_imb_pipeline(
        SMOTE(),
        model
    )

    model = pipeline.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    print("Evaluation on the test set:")
    test_accuracy = evaluate_model(y_test, y_pred)
    print("Evaluation on the training set:")
    train_accuracy = evaluate_model(y_train, y_train_pred)
    
    return train_accuracy, test_accuracy

```

```python
def evaluate_regression_model(x_train, x_test, y_train, y_test,model):
    X_train = x_train[:, :-1]
    X_test = x_test[:, :-1]
    pipeline = make_imb_pipeline(
        model
    )

    model = pipeline.fit(X_train, y_train)
    y_pred_test = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    
    # Tính toán và in ra MSE và RMSE cho tập kiểm tra và huấn luyện
    mse_test = mean_squared_error(y_test, y_pred_test)
    mse_train = mean_squared_error(y_train, y_pred_train)
    rmse_test = np.sqrt(mse_test)
    rmse_train = np.sqrt(mse_train)
    
    # Tính toán và in ra R^2 cho tập kiểm tra và huấn luyện
    r2_test = r2_score(y_test, y_pred_test)
    r2_train = r2_score(y_train, y_pred_train)
    
    # In kết quả
    dt= {
    "Metric": ["Mean Squared Error", "Root Mean Squared Error", "R-squared"],
    "Test Set": [mse_test, rmse_test, r2_test],
    "Train Set": [mse_train, rmse_train, r2_train]
    }
    dt = pd.DataFrame(dt)
    results_test = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_test, 'Time':x_test[:, -1]})
    results_train = pd.DataFrame({'Actual': y_train, 'Predicted': y_pred_train,'Time':x_train[:, -1] })
    return results_test,results_train ,dt
```

```python
def plot_evaluation_metrics(dataframes, models, metrics):

    fig, axes = plt.subplots(2, len(metrics), figsize=(18, 12))  

    for i, metric in enumerate(metrics):
        train_ax = axes[0, i] 
        test_ax = axes[1, i] 

        train_data = [df.loc[df['Metric'] == metric, 'Train Set'].iloc[0] for df in dataframes]
        test_data = [df.loc[df['Metric'] == metric, 'Test Set'].iloc[0] for df in dataframes]

        if metric == 'R-squared':
            value_function = max
        else:
            value_function = min

        best_train_value = value_function(train_data)
        best_test_value = value_function(test_data)
        
        train_colors = ['orange' if value == best_train_value else 'grey' for value in train_data]
        test_colors = ['orange' if value == best_test_value else 'grey' for value in test_data]
        
        # Vẽ biểu đồ
        train_ax.bar(models, train_data, color=train_colors, label='Train')
        test_ax.bar(models, test_data, color=test_colors, label='Test')

        train_ax.set_ylabel('Train ' + metric)
        test_ax.set_ylabel('Test ' + metric)
        train_ax.set_title(f'{metric} by Model')
        test_ax.set_title(f'{metric} by Model')
        train_ax.legend()
        test_ax.legend()

        # Hiển thị số liệu trên biểu đồ
        for j, model in enumerate(models):
            test_ax.text(j, test_data[j], f'{test_data[j]:.4f}', ha='center', va='bottom')
            train_ax.text(j, train_data[j], f'{train_data[j]:.4f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```

```python

models = ['RR', 'Gradient', 'SVR', 'DR', 'LR']
metrics = ['Mean Squared Error', 'Root Mean Squared Error', 'R-squared']
plot_evaluation_metrics(dataframes=[d_RR, d_Gradient, d_SVR, d_DR, d_LR], models=models, metrics=metrics)
```

```python
def data_vis(data,result, area = None):
    df = data.copy()
    df['forecastdate'] = pd.to_datetime(df['forecastdate'])
    df = df.merge(result, left_index=True, right_index=True)
    df = df.drop(columns='Time')
    df = df.loc[df['location'] == area]
    df['Actual'] = df['Actual'] -273.15
    df['Predicted'] = df['Predicted'] -273.15
    return df 
```
